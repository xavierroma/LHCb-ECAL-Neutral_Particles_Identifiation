```python
# 3. Import libraries and modules
import numpy as np
  # for reproducibility
np.random.seed(1)
from tensorflow import set_random_seed
set_random_seed(2)

import pandas as pd
from sklearn.model_selection import train_test_split
from keras.utils import plot_model
from keras.utils import to_categorical
from keras.models import Model
from keras.models import Sequential
from keras.layers import Input
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import Activation
from keras.losses import binary_crossentropy
from keras.losses import sparse_categorical_crossentropy
from keras.losses import categorical_crossentropy
from keras.layers.convolutional import Conv2D
from keras.layers.pooling import MaxPooling2D
from keras.layers.normalization import BatchNormalization
from keras.optimizers import SGD
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import metrics
from sklearn.metrics import roc_curve, auc
from scipy import interp
from sklearn.model_selection import StratifiedKFold
```

```python
#Load Data
photons_data = pd.read_csv('../data/MCUpgrade_photons.dat', sep='\t')
mergedpi_data = pd.read_csv('../data/MCUpgrade_mergedpi0.dat', sep='\t')
photons_data.columns = ['p','pt','eta','area','nPVs','ClusterE','ClusterE9','ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9','isPhr2','isPhr2r4','isPhasym','isPhkappa','isPhEseed','isPhE2']
mergedpi_data.columns = ['p','pt','eta','area','nPVs','ClusterE','ClusterE9','ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9','isPhr2','isPhr2r4','isPhasym','isPhkappa','isPhEseed','isPhE2']
```

```python
final_shape = ['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']
shape = ['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']
```

# Prepare Data..
- 0) Retrieve RAW data
- 1) Rotate RAW matrices
- 2) Normalise data
- 3) Add Original Variables to data
- 4) Add Labels to data
- 5) Merge both merged pi0 and photons data into single panda DataFrame -

```python
def rotate_to_top_matrice(m):
    max = 0
    if (m[0][1] > m[1][0]) & (m[0][1] > m[1][2]) & (m[0][1] > m[2][1]):
        max = 0
    elif (m[1][0] > m[1][2]) & (m[1][0] > m[2][1]):
        max = 1
    elif m[1][2] > m[2][1]:
        max = 2
    else:
        max = 3
    return np.rot90(m, max)
```

```python
# -0
m_d = pd.DataFrame()
p_d = pd.DataFrame()
data = pd.DataFrame()
#Retrive important columns
p_d = photons_data[['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']]
m_d = mergedpi_data[['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']]

# -1
# Rotate matrices to top direction
p_d = pd.DataFrame(np.array([rotate_to_top_matrice(m) for m in p_d.values.reshape((-1, 3, 3, 1))]).reshape(-1,9))
p_d.columns = ['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']
m_d = pd.DataFrame(np.array([rotate_to_top_matrice(m) for m in m_d.values.reshape((-1, 3, 3, 1))]).reshape(-1,9))
m_d.columns = ['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9'] 

# -2
#Normalise data
    #Divide each cell with the entire energy at the cluster
p_d = p_d[shape]
m_d = m_d[shape]
p_d = p_d.assign(suma = p_d.sum(axis=1));
m_d = m_d.assign(suma = m_d.sum(axis=1));
p_d_norm = (p_d.T / p_d.suma).T
m_d_norm = (m_d.T / m_d.suma).T

# -3
# Add to RAW matrice the original variables
m_d_norm[['isPhr2','isPhr2r4','isPhasym','isPhkappa','isPhEseed','isPhE2']] = mergedpi_data[['isPhr2','isPhr2r4','isPhasym','isPhkappa','isPhEseed','isPhE2']]
p_d_norm[['isPhr2','isPhr2r4','isPhasym','isPhkappa','isPhEseed','isPhE2']] = photons_data[['isPhr2','isPhr2r4','isPhasym','isPhkappa','isPhEseed','isPhE2']]

# -4
#Add Labels to data
m_d_norm['area'] = mergedpi_data[['area']]
p_d_norm['area'] = photons_data[['area']]

#Add category = 0 column to dataframe, so we identify merged pi0 as 0
m_d_norm = m_d_norm.assign(category=pd.Series(np.full((len(m_d_norm['area'])), 0)).values)

#Add category = 1 column to dataframe, so we identify photons as 1
p_d_norm = p_d_norm.assign(category=pd.Series(np.full((len(p_d_norm['area'])), 1)).values)

# -5
#Merge both mergedpi0 and photons into a single datagrame
data = m_d_norm.append(p_d_norm)
```

```python
#Function to divide data
#Param area [0,1,2] ECAL area
#test_fraction, the fraction of data to preserve for test
def divide_data(area, test_fraction):
    X_train, X_test, y_train, y_test = train_test_split(data[data['area'] == area][final_shape],
                                                data[data['area'] == area][['category']], test_size=test_fraction, random_state=0)
    return X_train.values.reshape((-1, 3, 3, 1)), X_test.values.reshape((-1, 3, 3, 1)), y_train, y_test
    
```

```python
def get_X_Y (area):
    X = data[data['area'] == area][final_shape].values.reshape((-1, 3, 3, 1))
    Y = data[data['area'] == area][['category']].values.reshape(-1)
    return np.array(X), np.array(Y)

```

```python
def print_model_history(hist):
    fig, axs = plt.subplots(1, 2, figsize=(20, 5), sharey=False)
    # Plot training & validation accuracy values
    axs[0].plot(hist.history['acc'])
    axs[0].plot(hist.history['val_acc'])
    axs[0].set_title('Model accuracy')
    axs[0].set_ylabel('Accuracy')
    axs[0].set_xlabel('Epoch')
    axs[0].legend(['Train', 'Test'], loc='upper left')


    # Plot training & validation loss values
    axs[1].plot(hist.history['loss'])
    axs[1].plot(hist.history['val_loss'])
    axs[1].set_title('Model loss')
    axs[1].set_ylabel('Loss')
    axs[1].set_xlabel('Epoch')
    axs[1].legend(['Train', 'Test'], loc='upper left')
    plt.show()
```

```python
def CV_CNN(build_model,X, Y, params, nsplits = 5):

    cv = StratifiedKFold(n_splits=nsplits, shuffle=True, random_state = 42)
    
    tprs = []
    aucs = []
    i = 0
    all_CNN = []
    all_predictions = []
    mean_fpr = np.linspace(0, 1, 100)
    for itrain, itest in cv.split(X, Y):
        X_train = np.array(X[itrain])
        X_test = np.array(X[itest])
        y_train = np.array(Y[itrain])
        y_test = np.array(Y[itest])
        
        all_CNN.append(build_model())
        
        history = all_CNN[i].fit(X_train, (y_train),
                        batch_size=params['batch_size'], epochs=params['epochs'], verbose=params['verbose'])

        all_predictions.append(all_CNN[i].predict(X_test))
        
        fpr, tpr, threshold = metrics.roc_curve(y_test, all_predictions[i])

        
        tprs.append(interp(mean_fpr, fpr, tpr))
        tprs[-1][0] = 0.0
        roc_auc = auc(fpr, tpr)
        aucs.append(roc_auc)
        
        plt.plot(fpr, tpr, lw=1, alpha=0.3, label='ROC fold %d (AUC = %0.2f)' % (i, roc_auc))
       
        i += 1

    mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    std_auc = np.std(aucs)
    

    
    return all_CNN, all_predictions, mean_fpr, mean_tpr, mean_auc, std_auc, plt

```

```python
def plot_results(mean_fpr, mean_tpr, mean_auc, std_auc, plot, color, title):
    plot.plot(mean_fpr, mean_tpr, color=color, label=r'Mean ROC (AUC = %0.2f $\pm$ %0.2f)' % (mean_auc, std_auc),
             lw=2, alpha=.8)
    plot.xlim([-0.01, 1.01])
    plot.ylim([-0.01, 1.01])
    plot.legend(loc = 'lower right')
    plot.ylabel('$\gamma$ efficiency')
    plot.xlabel('$\pi^0$ efficiency')
    plot.title(title)
    plot.show()

```

```python
def test_model(build_model):
    
    params = {'batch_size': 150, 
              'epochs': 50, 
              'verbose': 0}

    #OUTER
    X, Y = get_X_Y(0)
    
    plt.figure(figsize=(6,6))
    out_cnns, out_predictions, out_mean_fpr, out_mean_tpr, out_mean_auc, out_std_auc, out_plot = CV_CNN(build_model,X,Y,params,4)

    plot_results(out_mean_fpr, out_mean_tpr, out_mean_auc, out_std_auc, out_plot, 'r', "Outter Area; Original Data; CNN")
    
    #MIDDLE
    X, Y = get_X_Y(1)
    
    plt.figure(figsize=(6,6))
    mid_boosters, mid_predictions, mid_mean_fpr, mid_mean_tpr, mid_mean_auc, mid_std_auc, mid_plot = CV_CNN(build_model,X,Y,params,4)

    plot_results(mid_mean_fpr, mid_mean_tpr, mid_mean_auc, mid_std_auc, mid_plot, 'g', "Middle Area; RAW Data; CNN")

    #INNER
    X, Y = get_X_Y(2)

    plt.figure(figsize=(6,6))
    in_cnns, in_predictions, in_mean_fpr, in_mean_tpr, in_mean_auc, in_std_auc, in_plot =  CV_CNN(build_model,X,Y,params,4)

    plot_results(in_mean_fpr, in_mean_tpr, in_mean_auc, in_std_auc, in_plot, 'b', "Inner Area; Original Data; CNN")

    

    mean_fpr = np.linspace(0, 1, 100)
    tprs = []
    aucs = []

    tprs.append(interp(mean_fpr, in_mean_fpr, in_mean_tpr))
    tprs[-1][0] = 0.0
    aucs.append(in_mean_auc)

    tprs.append(interp(mean_fpr, out_mean_fpr, out_mean_tpr))
    tprs[-1][0] = 0.0
    aucs.append(out_mean_auc)

    tprs.append(interp(mean_fpr, mid_mean_fpr, mid_mean_tpr))
    tprs[-1][0] = 0.0
    aucs.append(mid_mean_auc)

    mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    std_auc = np.std(aucs)

    plt.figure(figsize=(10,10))

    plt.plot(out_mean_fpr, out_mean_tpr, color='r', label=r'Outter Area Mean ROC (AUC = %0.2f)' % (out_mean_auc),
                 lw=2, alpha=.8)
    plt.plot(mid_mean_fpr, mid_mean_tpr, color='g', label=r'Middle Area Mean ROC (AUC = %0.2f)' % (mid_mean_auc),
                 lw=2, alpha=.8)
    plt.plot(in_mean_fpr, in_mean_tpr, color='b', label=r'Inner Area Mean ROC (AUC = %0.2f)' % (in_mean_auc),
                 lw=2, alpha=.8)
    plt.plot(mean_fpr, mean_tpr, color='c', label=r'Mean ROC (AUC = %0.2f $\pm$ %0.2f)' % (mean_auc, std_auc),
                 lw=2, alpha=.8)

    plt.xlim([-0.01, 1.01])
    plt.ylim([-0.01, 1.01])
    plt.legend(loc = 'lower right')
    plt.title('CNN; RAW Data; Mean AUC')
    plt.ylabel('$\gamma$ efficiency')
    plt.xlabel('$\pi^0$ efficiency')





```

```python
import keras
def build_classifier(optimizer):
    model = Sequential()

    model.add(Conv2D(16, kernel_size=2, padding="same", kernel_initializer = 'glorot_uniform', input_shape=(3,3,1)))
    model.add(BatchNormalization())
    model.add(Activation("relu"))
    model.add(Flatten())
    model.add(Dense(224))
    model.add(BatchNormalization())
    model.add(Activation("relu"))
    model.add(Dense(112))
    model.add(BatchNormalization())
    model.add(Activation("relu"))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss=binary_crossentropy,
                  optimizer=optimizer,
                  
                  metrics=['accuracy'])
    return model
    
```

```python
def build_model1():
    model = Sequential()

    model.add(Conv2D(16, kernel_size=3, padding="same", kernel_initializer = 'glorot_uniform', input_shape=(3,3,1)))
    model.add(BatchNormalization())
    model.add(Activation("relu"))
    model.add(Flatten())
    model.add(Dense(224))
    model.add(BatchNormalization())
    model.add(Activation("relu"))
    model.add(Dense(112))
    model.add(BatchNormalization())
    model.add(Activation("relu"))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss=binary_crossentropy,
                  optimizer='sgd',
                  metrics=['accuracy'])
    return model

test_model(build_model1)
```

```python
def train_and_test_model(area, title, y_axis_label, x_axis_label, data_label, plt, color, show_info = False):
    X_train, X_test, y_train, y_test = divide_data(area,0.2)
    model = build_classifier('Adadelta')
    history = model.fit(X_train, (y_train.values),
                    batch_size=150, epochs=50, verbose=0, validation_data=(X_test, (y_test.values)))
    
    y_pred = model.predict(X_test)
    roc_auc = metrics.roc_auc_score(y_test.values, y_pred)
    
    false_positive_rate, true_positive_rate, thresholds = roc_curve((y_test.values), y_pred)
    plot_roc(false_positive_rate, true_positive_rate, roc_auc, title, y_axis_label, x_axis_label, data_label, plt,color)

    if show_info:
        print_model_history(history)
        print('Accuracy: %f' % model.evaluate(X_test, (y_test.values.reshape(-1, 1)), verbose=0)[1])
        
    return y_pred, y_test

plt.figure(figsize=(10,10))

y_pred_out, y_test_out = train_and_test_model(0, 'CNN - ROC Curves', '$\gamma$ efficiency','$\pi^0$ efficiency','Outter Area', plt, 'r--')
y_pred_mid, y_test_mid = train_and_test_model(1, 'CNN - ROC Curves', '$\gamma$ efficiency','$\pi^0$ efficiency','Middle Area',plt, 'g--')
y_pred_in, y_test_in = train_and_test_model(2, 'CNN - ROC Curves', '$\gamma$ efficiency','$\pi^0$ efficiency','Inner Area',plt, 'b--')

y_test = np.append(np.append(y_test_out, y_test_mid),y_test_in)
y_pred = np.append(np.append(y_pred_out, y_pred_mid),y_pred_in)

roc_auc = metrics.roc_auc_score(y_test, y_pred)
false_positive_rate, true_positive_rate, thresholds = roc_curve((y_test), y_pred)    
plot_roc(false_positive_rate, true_positive_rate, roc_auc, 'CNN - ROC Curves', '$\gamma$ efficiency','$\pi^0$ efficiency','Average', plt, 'c')

plt.show()
```
