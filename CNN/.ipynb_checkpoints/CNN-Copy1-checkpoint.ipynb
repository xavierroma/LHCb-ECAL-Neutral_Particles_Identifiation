```python
def get_data( raw = True ):
    photons_data = pd.read_csv('../data/MCUpgrade_photons.dat', sep='\t')

    mergedpi_data = pd.read_csv('../data/MCUpgrade_mergedpi0.dat', sep='\t')
    photons_data.columns = ['p','pt','eta','area','nPVs','ClusterE','ClusterE9','ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9','isPhr2','isPhr2r4','isPhasym','isPhkappa','isPhEseed','isPhE2']
    mergedpi_data.columns = ['p','pt','eta','area','nPVs','ClusterE','ClusterE9','ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9','isPhr2','isPhr2r4','isPhasym','isPhkappa','isPhEseed','isPhE2']
    mergedpi_data.head(10)
    m_d = pd.DataFrame()
    p_d = pd.DataFrame()
    data = pd.DataFrame()
    #Retrive important columns
    if raw:
        p_d = photons_data[['ClusterE9','ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']]
        m_d = mergedpi_data[['ClusterE9','ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']]
        #Normalise
        p_d = (p_d.T / p_d.ClusterE9).T[['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']]
        m_d = (m_d.T / m_d.ClusterE9).T[['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']]
        
    else:
        p_d = photons_data[['isPhr2','isPhr2r4','isPhasym','isPhkappa','isPhEseed','isPhE2']]
        m_d = mergedpi_data[['isPhr2','isPhr2r4','isPhasym','isPhkappa','isPhEseed','isPhE2']]
        
    m_d['area'] = mergedpi_data[['area']]
    p_d['area'] = photons_data[['area']]
    #Add category = 0 column to dataframe, so we identify merged pi0 as 0
    m_d = m_d.assign(category=pd.Series(np.full((len(m_d['area'])), 0)).values)

    #Add category = 1 column to dataframe, so we identify photons as 1
    p_d = p_d.assign(category=pd.Series(np.full((len(p_d['area'])), 1)).values)
    p_d_0 = p_d[p_d['area'] == 0]
    p_d_0 = p_d_0.iloc[0:len(m_d[m_d['area'] == 0])]

    p_d_1 = p_d[p_d['area'] == 1]
    p_d_1 = p_d_1.iloc[0:len(m_d[m_d['area'] == 1])]

    p_d_2 = p_d[p_d['area'] == 2]
    p_d_2 = p_d_2.iloc[0:len(m_d[m_d['area'] == 2])]

    #print (len(p_d_norm_0) + len(p_d_norm_1) + len(p_d_norm_2) == len(m_d))
    p_d = p_d_0.append(p_d_1).append(p_d_2)
    data = m_d.append(p_d)
    return data

```

```python
#2.4.6.8
def rotate_to_top_matrice(m):
    max = 0
    
    if (m[0][1] > m[1][0]) & (m[0][1] > m[1][2]) & (m[0][1] > m[2][1]):
        max = 0
    elif (m[1][0] > m[1][2]) & (m[1][0] > m[2][1]):
        max = 1
    elif m[1][2] > m[2][1]:
        max = 2
    else:
        max = 3
    return np.rot90(m, max)
        
    

#Function to divide data
#Param area [0,1,2] ECAL area
#test_fraction, the fraction of data to preserve for test
def divide_data(area, test_fraction, rotate = False):
    X_train, X_test, y_train, y_test = train_test_split(data[data['area'] == area][['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']],
                                                data[data['area'] == area][['category']], test_size=test_fraction, random_state=0)
    if rotate:
        X_train = np.array([rotate_to_top_matrice(m) for m in X_train.values.reshape((-1, 3, 3, 1))])
        X_test = np.array(([rotate_to_top_matrice(m) for m in X_test.values.reshape((-1, 3, 3, 1))]))
        return X_train, X_test, y_train, y_test
    
    return X_train.values.reshape((-1, 3, 3, 1)), X_test.values.reshape((-1, 3, 3, 1)), y_train, y_test
    
```

```python


#PProva 1
def build_classifier(optimizer):
    model = Sequential()

    model.add(Conv2D(16, kernel_size=3, padding="same", kernel_initializer = 'glorot_uniform', input_shape=(3,3,1)))
    model.add(BatchNormalization())
    model.add(Activation("relu"))
    model.add(Flatten())
    model.add(Dense(224))
    model.add(BatchNormalization())
    model.add(Activation("relu"))
    model.add(Dense(112))
    model.add(BatchNormalization())
    model.add(Activation("relu"))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss=binary_crossentropy,
                  optimizer=optimizer,
                  
                  metrics=['accuracy'])
    return model
    
```

```python
def plot_roc(false_positive_rate, true_positive_rate, roc_auc, title, y_axis_label, x_axis_label, data_label, plt,color):
    plt.title(title)
    plt.plot(false_positive_rate, true_positive_rate, color,
    label= '%s AUC = %0.2f'% (data_label, roc_auc))
    plt.legend(loc='lower right')
    plt.ylabel(y_axis_label)
    plt.xlabel(x_axis_label)
    #plt.show()
```

```python
def train_and_test_model(area, title, y_axis_label, x_axis_label, data_label, plt, color, show_info = False):
    X_train, X_test, y_train, y_test = divide_data(area,0.2)
    model = build_classifier('Adadelta')
    history = model.fit(X_train, (y_train.values),
                    batch_size=150, epochs=50, verbose=0, validation_data=(X_test, (y_test.values)))
    
    y_pred = model.predict(X_test)
    roc_auc = metrics.roc_auc_score(y_test.values, y_pred)
    
    false_positive_rate, true_positive_rate, thresholds = roc_curve((y_test.values), y_pred)
    plot_roc(false_positive_rate, true_positive_rate, roc_auc, title, y_axis_label, x_axis_label, data_label, plt,color)

    if show_info:
        print_model_history(history)
        print('Accuracy: %f' % model.evaluate(X_test, (y_test.values.reshape(-1, 1)), verbose=0)[1])
        
    return y_pred, y_test
    
```

```python
import tensorflow as tf
import keras.backend as K
import numpy as np


# -----------------------------------------------------------------------------------------------------------------------------------------------------
# AUC for a binary classifier
def auc(y_true, y_pred):
    ptas = tf.stack([binary_PTA(y_true, y_pred, k) for k in np.linspace(0, 1, 1000)], axis=0)
    pfas = tf.stack([binary_PFA(y_true, y_pred, k) for k in np.linspace(0, 1, 1000)], axis=0)
    pfas = tf.concat([tf.ones((1,)), pfas], axis=0)
    binSizes = -(pfas[1:] - pfas[:-1])
    s = ptas * binSizes
    return K.sum(s, axis=0)


# -----------------------------------------------------------------------------------------------------------------------------------------------------
# PFA, prob false alert for binary classifier
def binary_PFA(y_true, y_pred, threshold=K.variable(value=0.5)):
    y_pred = K.cast(y_pred >= threshold, 'float32')
    # N = total number of negative labels
    N = K.sum(1 - y_true)
    # FP = total number of false alerts, alerts from the negative class labels
    FP = K.sum(y_pred - y_pred * y_true)
    return FP / (N + 1)


# -----------------------------------------------------------------------------------------------------------------------------------------------------
# P_TA prob true alerts for binary classifier
def binary_PTA(y_true, y_pred, threshold=K.variable(value=0.5)):
    y_pred = K.cast(y_pred >= threshold, 'float32')
    # P = total number of positive labels
    P = K.sum(y_true)
    # TP = total number of correct alerts, alerts from the positive class labels
    TP = K.sum(y_pred * y_true)
    return TP / (P + 1)
```

```python
def auc_roc(y_true, y_pred):  
    # any tensorflow metric
    value, update_op = tf.metrics.auc(y_true, y_pred, num_thresholds=200, curve='ROC', summation_method='careful_interpolation' )
    # find all variables created for this metric
    metric_vars = [i for i in tf.local_variables() if 'auc_roc' in i.name.split('/')[1]]
    # Add metric variables to GLOBAL_VARIABLES collection.
    # They will be initialized for new session.
    for v in metric_vars:
        tf.add_to_collection(tf.GraphKeys.GLOBAL_VARIABLES, v)
    # force to update metric values
    with tf.control_dependencies([update_op]):
        value = tf.identity(value)
        return value
```
