```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
vmin = 0
vmax = 1
offset = 500
process = True
```

```python
#Load Data
photons_data = pd.read_csv('../data/MCUpgrade_photons.dat', sep='\t')
mergedpi_data = pd.read_csv('../data/MCUpgrade_mergedpi0.dat', sep='\t')
photons_data.columns = ['p','pt','eta','area','nPVs','ClusterE','ClusterE9','ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9','isPhr2','isPhr2r4','isPhasym','isPhkappa','isPhEseed','isPhE2']
mergedpi_data.columns = ['p','pt','eta','area','nPVs','ClusterE','ClusterE9','ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9','isPhr2','isPhr2r4','isPhasym','isPhkappa','isPhEseed','isPhE2']
```

```python
# Original + Raw
# final_shape = ['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9','isPhr2','isPhr2r4','isPhasym','isPhkappa','isPhEseed','isPhE2']
#Raw
final_shape = ['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']
# Original
#final_shape = ['isPhr2','isPhr2r4','isPhasym','isPhkappa','isPhEseed','isPhE2']
shape = ['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']
```

```python
# -0
m_d = pd.DataFrame()
p_d = pd.DataFrame()
data = pd.DataFrame()
#Retrive important columns
p_d = photons_data[['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']]
m_d = mergedpi_data[['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']]
```

```python
def rotate_to_top_matrice(m):
    max = 0
    if (m[0][1] > m[1][0]) & (m[0][1] > m[1][2]) & (m[0][1] > m[2][1]):
        max = 0
    elif (m[1][0] > m[1][2]) & (m[1][0] > m[2][1]):
        max = 3
    elif m[1][2] > m[2][1]:
        max = 1
    else:
        max = 2
    return np.rot90(m, max)



    

```

```python
if (process):
    # -1
    # Rotate matrices to top direction
    p_d = pd.DataFrame(np.array([rotate_to_top_matrice(m) for m in p_d.values.reshape((-1, 3, 3, 1))]).reshape(-1,9))
    p_d.columns = ['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']
    m_d = pd.DataFrame(np.array([rotate_to_top_matrice(m) for m in m_d.values.reshape((-1, 3, 3, 1))]).reshape(-1,9))
    m_d.columns = ['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']

    # -2
    #Normalise data
        #Divide each cell with the entire energy at the cluster
    p_d = p_d[shape]
    m_d = m_d[shape]
    p_d = p_d.assign(suma = p_d.sum(axis=1));
    m_d = m_d.assign(suma = m_d.sum(axis=1));
    p_d_norm = (p_d.T / p_d.suma).T
    m_d_norm = (m_d.T / m_d.suma).T

    # -4
    #Add Labels to data
    m_d_norm['area'] = mergedpi_data[['area']]
    p_d_norm['area'] = photons_data[['area']]

    #Add category = 0 column to dataframe, so we identify merged pi0 as 0
    m_d_norm = m_d_norm.assign(category=pd.Series(np.full((len(m_d_norm['area'])), 0)).values)

    #Add category = 1 column to dataframe, so we identify photons as 1
    p_d_norm = p_d_norm.assign(category=pd.Series(np.full((len(p_d_norm['area'])), 1)).values)


    # -5 
    # Equilibrate category classes
    p_d_norm_0 = p_d_norm[p_d_norm['area'] == 0]
    p_d_norm_0 = p_d_norm_0.iloc[0:len(m_d_norm[m_d_norm['area'] == 0])]

    p_d_norm_1 = p_d_norm[p_d_norm['area'] == 1]
    p_d_norm_1 = p_d_norm_1.iloc[0:len(m_d_norm[m_d_norm['area'] == 1])]

    p_d_norm_2 = p_d_norm[p_d_norm['area'] == 2]
    p_d_norm_2 = p_d_norm_2.iloc[0:len(m_d_norm[m_d_norm['area'] == 2])]

    print (len(p_d_norm_0) + len(p_d_norm_1) + len(p_d_norm_2) == len(m_d_norm))
    p_d_norm = p_d_norm_0.append(p_d_norm_1).append(p_d_norm_2)
    p_d = p_d_norm
    m_d = m_d_norm
    vmin = 0
    vmax = 1
    offset = 500
else:
    m_d['area'] = mergedpi_data[['area']]
    p_d['area'] = photons_data[['area']]
    
    vmin = 0
    vmax = 70000
    offset = 100

```

```python

x = [i for range_set in range(3) for i in range (1,4)]
y = sorted(x, reverse=True)
  
single_m_0 = np.array(m_d[m_d['area'] == 0][['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']])[0]
single_m_1 = np.array(m_d[m_d['area'] == 1][['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']])[0]
single_m_2 = np.array(m_d[m_d['area'] == 2][['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']])[1]

single_p_0 = np.array(p_d[p_d['area'] == 0][['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']])[0]
single_p_1 = np.array(p_d[p_d['area'] == 1][['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']])[0]
single_p_2 = np.array(p_d[p_d['area'] == 2][['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']])[0]

mean_m_0 = (m_d[m_d['area'] == 0][['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']].values).mean(axis=0)
mean_m_1 = (m_d[m_d['area'] == 1][['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']].values).mean(axis=0)
mean_m_2 = (m_d[m_d['area'] == 2][['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']].values).mean(axis=0)

mean_p_0 = (p_d[p_d['area'] == 0][['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']].values).mean(axis=0)
mean_p_1 = (p_d[p_d['area'] == 1][['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']].values).mean(axis=0)
mean_p_2 = (p_d[p_d['area'] == 2][['ClE1','ClE2','ClE3','ClE4','ClE5','ClE6','ClE7','ClE8','ClE9']].values).mean(axis=0)



def plot_cluster_comparasion(weight_0, weight_1):
    f, (ax) = plt.subplots(1, 2, figsize=(12,5),sharex=True, sharey=True, constrained_layout=True)
    levels = np.linspace(0, 1)
    im = ax[0].hist2d(x, y, (3, 3), weights=weight_0, cmap=plt.cm.jet, vmin=vmin, vmax=vmax)
    ax[1].hist2d(x, y, (3, 3), weights=weight_1, cmap=plt.cm.jet,vmin=vmin, vmax=vmax)
   
    plt.colorbar(im[3]);
    
    
    plt.show()




```

```python

plot_cluster_comparasion(mean_m_0, mean_p_0)
plot_cluster_comparasion(mean_m_1, mean_p_1)
plot_cluster_comparasion(mean_m_2, mean_p_2)

```

```python

plot_cluster_comparasion(single_m_0, single_p_0)
plot_cluster_comparasion(single_m_1, single_p_1)
plot_cluster_comparasion(single_m_2, single_p_2)

```
